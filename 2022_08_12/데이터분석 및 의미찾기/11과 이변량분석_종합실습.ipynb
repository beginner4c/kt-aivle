```python
path = 'https://raw.githubusercontent.com/JunhoYu/Data-Analysis/main/Carseats.csv'
data = pd.read_csv(path)
data.drop(['Income', 'Urban', 'Education'],axis = 1, inplace = True)
data.head()
```

```python
target = 'Sales'
```

### (1) 숫자형(수치형) X --> Y
* 모든 숫자형 X에 대해서 Y와 비교하여 차트를 그리고 수치화 하시오.

① 시각화 : scatter, jointplot  
② 수치화 : 상관분석

#### ① Advertising -> Sales

1) 시각화 : scatter(regplot), jointplot

```python
var = 'Advertising'

sns.scatterplot(x=var, y=target, data=data) # 산점도
plt.show()
```

```python
sns.regplot(x=var, y = target, data=data) # 산점도와 함께 시각화를 도와줄 직선을 그려준다
plt.show()
```

```python
sns.jointplot(x=var, y=target, data=data)
plt.show()
```

2) 수치화 : 상관분석

```python
result = spst.pearsonr(data[var], data[target])
print(f"상관계수 : {result[0]}, p-value : {result[1]}")
```
